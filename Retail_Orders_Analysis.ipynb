# Create New Columns: Discount, Sales Price, Profit

```python
# discount column
df['discount']=df['list price']*df['discount percent']*.01
df
```

```python
# sale price column
df['sale price'] = df['list price']-df['discount']
df
```

```python
# profit column
df['profit'] = df['sale price']-df['cost price']
df
```
